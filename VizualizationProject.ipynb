## Define Panel widgets

```python
year_slider = pn.widgets.IntSlider(name='Covid19 cases untill the year of', start=2021, end=2023, step=1, value=2022)
year_slider
```

```python
month_slider = pn.widgets.IntSlider(name='Untill month', start=1, end=12, step=1, value=6)
month_slider
```

### Viz1- Radio buttons for new and total cases of COVID-19 per million over time by continent

```python
yaxis_new_cases = pn.widgets.RadioButtonGroup(name='year', 
    options=['total_cases_per_million', 'new_cases_per_million'],
    button_type='success')
```

```python
continents = ['Asia', 'Unknown' , 'Europe', 'Africa', 'North America', 'South America', 'Oceania']

new_cases_pipeline = (i_df[(i_df['year'] <= year_slider) & (i_df['month'] <= month_slider) & (i_df.continent.isin(continents))]
    .groupby(['continent', 'year'])[yaxis_new_cases].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True))
```

```python
new_cases_pipeline
```

```python
new_cases_plot = new_cases_pipeline.hvplot(x = 'year', by='continent', y = yaxis_new_cases ,line_width=2, title="Cases per million over time by continent")
new_cases_plot
```

### Viz2 - Table for new and total cases of COVID-19 per million over time by continent   

```python
new_and_total_cases_table = new_cases_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 13, sizing_mode='stretch_width', layout='fit_data', width=450) 
new_and_total_cases_table
```

### Viz3- Scatter plot for new deaths VS new cases of COVID-19 smoothed per million by countries in Asia

```python
locations = ['Hong Kong', 'Georgia', 'South Korea', 'Israel', 'Brunei', 'Maldives', 'Taiwan', 'Vietnam', 'Lebanon', 'Kyrgyzstan', 'Bahrain', 'Turkey']
```

```python
new_deaths_vs_new_cases_scatterplot_pipeline = (i_df[(i_df.continent.isin(continents)) & (i_df['continent'] == 'Asia') & (i_df.location.isin(locations))]
    .groupby(['location', 'year', 'new_deaths_smoothed_per_million'])['new_cases_smoothed_per_million'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True))
```

```python
new_deaths_vs_new_cases_scatterplot_pipeline
```

```python
new_deaths_vs_new_cases_scatterplot = new_deaths_vs_new_cases_scatterplot_pipeline.hvplot(x='new_cases_smoothed_per_million', 
                                                                y='new_deaths_smoothed_per_million', by='location', size=80, kind="scatter", 
                                                                alpha=0.7, legend=True, title='New cases Vs New deaths by countries in Asia',sizing_mode='fixed', height=500, width=800)
new_deaths_vs_new_cases_scatterplot
```

```python
yaxis_death_cases = pn.widgets.RadioButtonGroup(
    name='location', 
    options=['total_deaths_per_million', 'new_deaths_per_million'], 
    button_type = 'success')

death_cases_bar_pipeline = (
    i_df[(i_df.continent.isin(continents))]
    .groupby(['year', 'continent'])[yaxis_death_cases].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True))
```

```python
death_cases_bar_pipeline
```

```python
yaxis_death_cases_bar_plot = death_cases_bar_pipeline.hvplot(kind='bar', x='continent', y=yaxis_death_cases, title='Total Deaths Vs New deaths by continent', width=600)
yaxis_death_cases_bar_plot
```
